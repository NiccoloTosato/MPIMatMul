```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('time.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=' ')
    line_count = 0
    values=list()
    for row in csv_reader:
            values.append(row)
            line_count += 1
values=np.array(values,dtype=np.float32)

displ=np.array([0,1,2,3,4])
nodes=np.array([1,2,4,8,16])
plt.figure(figsize=(10,10))
plt.bar(displ,values[:,1],width=0.25,label="MPI",alpha=0.8,color='r')
plt.bar(displ+0.27,values[:,2],width=0.25,label="Homemade",alpha=0.8,color='b')

plt.xlabel("Nodes")
plt.ylabel("Time [s]")

plt.xticks(ticks=[0,1,2,3,4],labels=map(int,nodes))
plt.xticks(ticks=[0,1,2,3,4],labels=map(int,nodes))
plt.legend()
plt.title(r"FFTW N=$512^3$")
plt.show()
```
