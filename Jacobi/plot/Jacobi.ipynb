```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open('cpu.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=' ')
    line_count = 0
    values=list()
    for row in csv_reader:
            values.append(row)
            line_count += 1
values=np.array(values,dtype=np.float32)

with open('gpu.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=' ')
    line_count = 0
    values_cuda=list()
    for row in csv_reader:
        values_cuda.append(row)
        line_count += 1
values_cuda=np.array(values_cuda,dtype=np.float32)
displ=np.array([0,1,2,3,4])
nodes=np.array([1,2,4,8,16])
plt.figure(figsize=(10,10))
plt.bar(displ,values[:,1],width=0.25,label="Compute",alpha=0.8,color='r')
plt.bar(displ,values[:,2],bottom=values[:,1],width=0.25,label="Communication",alpha=0.8,color='b')

plt.bar(displ+0.27,values_cuda[:,1],width=0.25,alpha=0.8,color='r', hatch='///')
plt.bar(displ+0.27,values_cuda[:,2],bottom=values_cuda[:,1],width=0.25,alpha=0.8,color='b', hatch='///')
plt.bar([0],[0],fill=False, hatch='///',label='GPU')
plt.bar([0],[0],fill=False,label='CPU')
plt.xlabel("Nodes")
plt.ylabel("Time [s]")

plt.xticks(ticks=[0,1,2,3,4],labels=map(int,nodes))
plt.xticks(ticks=[0,1,2,3,4],labels=map(int,nodes))
plt.legend()
plt.title("Jacobi N25000")
plt.show()
```
